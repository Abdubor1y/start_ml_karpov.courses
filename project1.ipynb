```python
unique_playstore = playstore.groupby("App", as_index=False).first()
unique_playstore
```

```python
playstore.columns = (
    playstore.columns
    .str.lower()
    .str.replace(" ", "_")
)

playstore
```

```python
share = (unique_playstore["Type"] == "Free").mean()
print(round(share, 2))
```

```python
playstore['reviews'] = pd.to_numeric(playstore['reviews'])

education_playstore = playstore[
    (playstore['category'] == 'EDUCATION') & (playstore['reviews'] > 1000)
]

education_playstore.reset_index(drop=True)
```

```python
replace_symb_to_float = lambda x: float(x.replace('$', ''))
unique_playstore.price = unique_playstore.price.apply(replace_symb_to_float)

unique_playstore.head()
```

```python
price_median = round(unique_playstore.price.median(), 2)
price_mean = round(unique_playstore.price.mean(), 2)
print(price_median, price_mean)
```
